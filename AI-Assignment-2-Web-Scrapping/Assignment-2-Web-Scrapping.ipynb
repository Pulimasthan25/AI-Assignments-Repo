```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://realpython.github.io/fake-jobs/"
```

```python
response=requests.get(url)
```

```python
response
```

```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
job_title=[]
company_name=[]
date=[]
```

```python
title=soup.findAll("div",attrs={"class":"column is-half"})
```

```python
for store in title:
    name=store.h2.text
    job_title.append(name)
    company=store.h3.text
    company_name.append(company)
    posted=store.div.find("p",class_="is-small has-text-grey").text.replace("\n","")
    date.append(posted)
```

```python
job_dataSet=pd.DataFrame({"Job Title":job_title,"Company Name":company_name,"Date Posted":date})
job_dataSet
```
